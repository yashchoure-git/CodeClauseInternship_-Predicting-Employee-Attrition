```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("employee_attrition_dataset.csv")
dataset.head()
```

```python
dataset['Age'].unique()
```

```python
dataset['Gender'].unique()
```

```python
dataset["Gender"]=dataset["Gender"].astype(str).str.strip().str.capitalize()
```

```python
dataset["Gender"]=dataset["Gender"].map({'Male':1,'Female':0})
```

```python
dataset['Department'].unique()
```

```python
dataset['Department']=dataset['Department'].map({'Sales':0,'IT':1,'Finance':2,'Operations':3,'HR':4})
```

```python
dataset=dataset.drop(columns=["Overtime"])
```

```python
dataset['JobRole'].unique()
```

```python
dataset['JobRole']=dataset['JobRole'].map({'Manager':0, 'Sales Rep':1, 'Executive':2, 'Engineer':3, 'Analyst':4})
```

```python
dataset['Attrition'].unique()
```

```python
dataset['Attrition']=dataset['Attrition'].map({'Yes':1, 'No':0})
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['Attrition']
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rs=RandomUnderSampler()
rs_x,rs_y=rs.fit_resample(x,y)
```

```python
rs_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
gnb=LogisticRegression()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```
